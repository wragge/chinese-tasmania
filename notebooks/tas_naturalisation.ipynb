```python
import re
import pandas as pd
from pathlib import Path
```

```python
df = pd.read_csv("https://data.gov.au/data/dataset/b19224c6-3b78-4d03-8a9d-c97d7b8db541/resource/6e4a4678-9845-42da-a7e1-99f00aee2028/download/naturalisations.csv")
```

```python
Path("nat_places.txt").write_text("\n".join(sorted(list(df.dropna(subset=["SHIP_NATIVE_PLACE"])["SHIP_NATIVE_PLACE"].unique()))))
```

```python
df_place = df.dropna(subset=["SHIP_NATIVE_PLACE"])
df_china = df_place.loc[df_place["SHIP_NATIVE_PLACE"].str.contains("china|hong kong", regex=True, flags=re.IGNORECASE,)].copy()
```

```python
names = list(df_china["NAME"].unique())
```

```python
name_parts = []
for name in names:
    name_parts += re.findall(r"\w+", name)
    
name_parts = [n for n in list(set(name_parts)) if len(n) > 1]
```

```python
all_name_parts = []
for name in list(df["NAME"].unique()):
    all_name_parts += re.findall(r"\w+", name)
    
#all_name_parts = [n for n in list(set(name_parts)) if len(n) > 1]
```

```python
from collections import Counter
name_counts= Counter(all_name_parts)
name_counts.most_common(50)
most_common = [n[0] for n in name_counts.most_common(50)]
```
